```python
import pandas as pd
import matplotlib.pyplot as plt

result = df.groupby('jenis_bencana')['jumlah_korban_bencana'].sum()

result = result.sort_values(ascending=False)


result.plot(kind='bar')
plt.title('Jumlah Korban  Bencana ')
plt.xlabel('Kategori Bencana')
plt.ylabel('Jumlah Korban')
plt.show()
```

```python
total_korban_per_kategori = df.groupby('jenis_bencana')['jumlah_korban_bencana'].sum()
kategori_tertinggi = total_korban_per_kategori.idxmax()

print(f"Kategori bencana dengan jumlah korban tertinggi: {kategori_tertinggi}")
```
